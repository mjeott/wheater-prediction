```python
X = prediksi_cuaca_dataset.drop(columns='output', axis=1)
Y = prediksi_cuaca_dataset['output'] 
```

Standarisasi data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X) 
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = prediksi_cuaca_dataset['output'] 
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi data training adalah =', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Akurasi data testing adalah =', test_data_accuracy)
```

```python
import numpy as np
# Data input
input_data = (2.5, 1.7, -2.8, 5)

# Memuat model yang telah dilatih
from sklearn.linear_model import LogisticRegression  # Misalkan classifier menggunakan LogisticRegression

# Memuat scaler yang telah difit
from sklearn.preprocessing import StandardScaler

# Ubah input_data menjadi array numpy
input_data_as_numpy_array = np.array(input_data)

# Ubah bentuk array menjadi (1, -1)
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

# Standarisasi data
std_data = scaler.transform(input_data_reshape)

# Lakukan prediksi
prediction = classifier.predict(std_data)

# Interpretasi hasil prediksi
if prediction[0] == 0:
    print('Aman untuk keluar')
elif prediction[0] == 1:
    print('Aman untuk keluar')
elif prediction[0] != 2:
    print('Tidak aman untuk keluar')
elif prediction[0] != 3:
    print('Tidak aman untuk keluar')
else:
    print('Terjadi kesalahan pada model. Silakan periksa kembali model Anda.')
```
